# Tratando os dados para pegar apenas até 2022

Os dados foram coletados em 17/03/2023, assim já contém os dados do GP de 05/03/2023. Por isso, esse notebook tem como objetivo limpar os dados e garantir que contém dados até 2022 apenas.

```python
import pandas as pd 
```

## circuits.csv

```python
circuits = pd.read_csv("data/circuits.csv")
circuits
```

```python
circuits.to_csv("processed_data/circuits.csv")
```

## races.csv

```python
races = pd.read_csv("data/races.csv")
races

races_2023 = races.loc[races.year == 2023]

processed_races = races.loc[races.year != 2023]
processed_races
```

```python
processed_races.to_csv("processed_data/races.csv")
```

## constructor_results.csv

```python
constructor_results = pd.read_csv("data/constructor_results.csv")
constructor_results
processed_constructor_results = constructor_results.loc[~constructor_results.raceId.isin(races_2023.raceId.tolist())]
```

```python
processed_constructor_results.to_csv("processed_data/constructor_results.csv")
```
